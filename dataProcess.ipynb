```python
import polars as pl

data_path = '../data/returnForcast.parquet'
data = pl.read_parquet(data_path).to_pandas()
print(data.head())
```

```python
import pandas as pd

macro_data_path = '../data/macroFactor.csv'
zero_rate_data_path = '../data/zeroRate.csv'

macro_data = pd.read_csv(macro_data_path)
zero_rate_data = pd.read_csv(zero_rate_data_path)

print(macro_data.head())
print(zero_rate_data.head())
print(macro_data.columns)
```

```python
# 步骤A: 删除 date 为空的行，防止 NaN 导致类型强制为 float
data = data.dropna(subset=['date','ret'])
print(data.shape)

macro_data.rename(columns={'month': 'date'}, inplace=True)
zero_rate_data.rename(columns={'month': 'date'}, inplace=True)

# print(data.head())
print(macro_data.head())
print(zero_rate_data.head())
```

```python
# 步骤B: 合并数据
data = data.merge(macro_data, on='date', how='left')
data = data.merge(zero_rate_data, on='date', how='left')

print(data.head())
```

```python
print(data.shape)

data['original_ret'] = data['ret']
data['ret'] = data['ret'] - data['RF']

data.to_parquet('../data/returnForcast_Macro.parquet', index=False)
```
